```python
import csv
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

# 查看数据各列的信息

```python
data = pd.read_csv(r'vgsales.csv')
data.head(10)
```

# 给出销量前10的游戏并可视化

```python
Top_Games = data[['Name','Global_Sales']]
Top_Games = Top_Games.sort_values('Global_Sales', ascending = False)
```

```python
Top_Games = Top_Games[:10]
names = list(Top_Games['Name'])
Sales = list(Top_Games['Global_Sales'])
```

```python
plt.figure(figsize=(20, 8), dpi=80)
x = range(len(names))
plt.bar(x, Sales, width = 0.6)
plt.xticks(x, names)
```

# 统计销量最高的游戏类型，并可视化

```python
Genre_sales = data[['Genre', 'Global_Sales']]
```

```python
Genre = data['Genre']
```

```python
genre_table = {}
for e in Genre_sales.itertuples():
    if e[1] not in genre_table.keys():
        genre_table[e[1]] = e[2]
    else:
        genre_table[e[1]] += e[2]
```

```python
genre_table = list(genre_table.items())
genre_table.sort(key = lambda x: x[1], reverse = True)
```

```python
genre_table
```

```python
genres = [i[0] for i in genre_table]
y = [i[1] for i in genre_table]
x = range(len(genres))
plt.figure(figsize = (20, 8), dpi = 80)
plt.bar(x, y, width = 0.6)
plt.xticks(x, genres)
```
